```python
# these are the apckages we need 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mlp
from scipy.stats import linregress 
```

What plots do we want..

we have data for the five counties. 



```python
mergedData = pd.read_csv('resources/merged_weather.csv')
wineData = pd.read_csv('resources/Californa_Wine_Production_1980_2020.csv')
```

```python
wineDatasm = wineData[wineData['County'].isin(['Placer','Sonoma','Napa','SantaCruz','Marin'],)]
```

```python
plt.scatter(wineData['Year'], wineData['Production'], c=wineData['CountyCode'])

plt.title("Production by Year by County [ALL]")
plt.xlabel("Year")
plt.ylabel("Production")

# Display the plot
plt.show()

plt.scatter(wineDatasm['Year'], wineDatasm['Production'], c=wineDatasm['CountyCode'])

plt.title("Production by Year by County [Top 5]")
plt.xlabel("Year")
plt.ylabel("Production")

# Display the plot
plt.show()
```

we can see a great discrepensy in production by county, even with the top five counties. this may be due to the size of the counties themselves. to control for size differences. we can plot via yield (tons per acres).

```python

colors = ['red', 'black', 'green', 'blue', 'purple']

for (color, county) in zip(colors, wineDatasm['County'].unique()):
    filtered = wineDatasm[wineDatasm['County']== county]
    plt.scatter(filtered['Year'], filtered['Yield(Unit/Acre)'], c=color, label = county)

    (slope, intercept, rvalue, pvalue, stderr) = linregress(filtered['Year'], filtered['Yield(Unit/Acre)'])
    predicted_y = slope*filtered['Year'] + intercept
    _ = plt.plot(filtered['Year'], predicted_y, color = color)

# plot1 = plt.scatter(wineDatasm['Year'], wineDatasm['Yield(Unit/Acre)'], c=wineDatasm['CountyCode'])

plt.title("Yield(Unit/Acre) by Year by County [Top 5]")
plt.xlabel("Year")
plt.ylabel("Yield(Unit/Acre)")
# plot1 = plt.scatter(wineDatasm['Year'], wineDatasm['Yield(Unit/Acre)'], c=wineDatasm['CountyCode'])

# plt.title("Yield(Unit/Acre) by Year by County [Top 5]")
# plt.xlabel("Year")
# plt.ylabel("Yield(Unit/Acre)")
# plt.legend() # add legend here


# need to add legend for counties
plt.legend()
plt.show()
```
